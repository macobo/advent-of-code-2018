## --- Part Two ---
Good thing you didn't have to wait, because that would have taken a long time - much longer than the 3 seconds in the example above.

Impressed by your sub-hour communication capabilities, the Elves are curious: exactly how many seconds would they have needed to wait for that message to appear?

Your puzzle answer was 10888.

````python
import re
parse = lambda line: list(map(int, re.findall(r"(-?\d+)", line)))
data = list(map(parse, unparsed_data))
````

````python
from collections import defaultdict
grid = defaultdict(int)

def position(time, initial_point):
    x, y, dx, dy = initial_point
    return (x + time * dx, y + time * dy)

def tick(time, data):
    return [position(time, point) for point in data]

tick(19000, data)
````

````python
def distance(positions):
    sorted_positions = list(sorted(positions))
    return sorted_positions[-1][0] - sorted_positions[0][0]

min(range(20000), key = lambda time: distance(tick(time, data)))
````

````python
def minmax(iterator):
    llist = list(iterator)
    return min(llist), max(llist)

def present(data):
    grid = defaultdict(lambda: ' ')
    for xy in data: grid[xy] = '#'
    
    min_x, max_x = minmax(x for x, _ in data)
    min_y, max_y = minmax(y for _, y in data)
    
    print('\n'.join([''.join(grid[(x, y)] for x in range(min_x, max_x + 1)) for y in range(min_y, max_y + 1)]))
````

````python
present(tick(10888, data))
````
